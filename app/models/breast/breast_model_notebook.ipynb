```python
plt.figure(figsize = (30, 25))
sns.heatmap(df.corr(), annot = True);
```

```python
plt.figure(figsize=(8, 6))

corr_by_winner = sns.heatmap(
    df.corr()[["target"]].sort_values(by="target", ascending=False),
    vmin=-1,
    vmax=1,
    cmap="BrBG",
    annot=True,
)

corr_by_winner.set_title(
    "Target x Features", fontdict={"fontsize": 20}, pad=16
)

plt.show()
```

```python
fig, axs = plt.subplots(6, 5, figsize=(20, 20))

axs = axs.flatten()

for i, (coluna, ax) in enumerate(zip(df.columns, axs)):
    ax.boxplot(df[coluna])
    ax.set_title(coluna)

plt.tight_layout()
plt.show()
```

```python
for col_name in df.columns:
    q1 = df[col_name].quantile(0.25)
    q3 = df[col_name].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    df[col_name] = np.where((df[col_name] > upper_bound) | (df[col_name] < lower_bound), np.mean(df[col_name]), df[col_name])
```

```python
fig, axs = plt.subplots(6, 5, figsize=(20, 20))

axs = axs.flatten()

for i, (coluna, ax) in enumerate(zip(df.columns, axs)):
    ax.boxplot(df[coluna])
    ax.set_title(coluna)

plt.tight_layout()
plt.show()
```
